```python
#remove duplicates in list
num=[2,2,3,4,5,1,1,6,2,7]
rem=[]
for i in num:
  if i not in rem:
    rem.append(i)
print("Actual list: ",num)
print("After removal: ",rem)
```

```python
#find all pairs in a list whose sum is equal to target
num=[2,4,3,5,6,7,8,9]
target=10
rem=[]
for i in range (len(num)): #range (len(num))
  for j in range(i+1, len(num)):
    if (num[i]+num[j]==target):
      rem.append((num[i],num[j]))
print("pair is: ",rem)
```

```python
# find frequency of each number
num=[1,2,2,3,3,4,4,4,5,5,5]
dict={}
for i in num:
  dict[i]=dict.get(i,0)+1
print(dict)

dict2={}
for i in dict2:
  dict2[i]+=1
else:
  dict2[i]=1
print(dict2)
```
