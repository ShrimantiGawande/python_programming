```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

#importing dataset
sh=np.array([1,2,3,4,5,6,7]).reshape(-1,1)
sg=np.array([8,9,10,11,12,13,14])

#convert linear input to polynomial feature
#degree 2 means model will learn x
poly=PolynomialFeatures(degree=2)
sh_poly=poly.fit_transform(sh)

model=LinearRegression()
model.fit(sh_poly,sg)
sg_pred=model.predict(sh_poly)

print("intercept",model.intercept_)
print("coefficient of sh(b)",model.coef_[0])
print("coef of sh^2(a)", model.coef_[1])

#visualization
plt.scatter(sh,sg,color="red")
plt.plot(sh,sg_pred,color="blue")
plt.shlabel("sh")
plt.sglabel("sg")
plt.legend()
plt.show
```
