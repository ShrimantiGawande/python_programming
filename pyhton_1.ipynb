```python
#find largest of 3 numbers
a,b,c=12,36,76
if a>=b & a>=c:
  print("a largest number")
if b>=a & b>=c:
  print("b is largest")
else:
  print("c is largest")
```

```python
#if numbers are same
a,b,c=12,76,76
if a>=b & a>=c:
  print("a largest number")
elif b>=a & b>=c:
  print("b is largest")
if c>=a & c>=b:
  print("c is largest")
```
