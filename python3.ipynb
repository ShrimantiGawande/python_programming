```python
import numpy as np
shri=np.linspace(0,10,50)
shrim=shri*5
shrimp=shri*2
shrlab=["raj", "shri", "roy", "mona", "ashish"]
sg=np.random.randint(0,10,5)
plt.figure()
plt.plot(shrim, shrimp, color="red", label="dataset")
plt.grid(True)
plt.xlabel("number of people")
plt.ylabel("arrange marks")
plt.title("spread of average score")
plt.legend()

color=["red", "blue", "black", "green", "grey"]
plt.pie(sg, labels=shrlab, autopct="%1.1f%%")
plt.title("spread of average score")
plt.show()
```

```python
# Q.1] Plot a histogram for any numerical column from a dataset of your choice and interpret the distribution.
import matplotlib.pyplot as plt
import pandas as pd
sg = {
"sr no" : [1,2,5,6,7,8],
"income in thousands" : [97,96,85,90,87,75]
}
ahiru = pd.DataFrame(sg)
plt.figure()
plt.hist(ahiru['sr no'],bins = 5)
plt.xlabel("sr no")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
# Q.2] Create a scatter plot between two numerical features and comment on their relationship.
plt.figure()
plt.scatter(ahiru['sr no'],ahiru['income in thousands'])
plt.xlabel("Sr No")
plt.ylabel("Income")
plt.title("Scatter Plot")
plt.show()
# There is no best fit line in these points.
```

```python
# Q.3] Compute the correlation matrix for a dataset and identify the most correlated feature pair.
correlation_matrix = ahiru.corr()
print("Correlation Matrix : ",correlation_matrix)
```

```python
# Q.4] Use a box plot to identify outliers in a numerical column.
plt.figure()
plt.boxplot(ahiru['income in thousands'])
plt.ylabel("Income")
plt.title("Box Plot of Marks")
plt.show()
```
